```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def MMR_super_period(p_transit, p_other, j, k):
    '''
    from eq 9, here https://arxiv.org/pdf/1706.09849.pdf
    '''
    
    p_ttv = 1 / np.abs((j / p_other) - (k / p_transit))

    
    
    return p_ttv
```

```python
MMR_super_period(160.9, 320, 2, 1)
```

```python
p_other = np.arange(60, 900, .1)
```

```python
p_koi3678 = 160.8845
mmr_periods = []
ratios = []
ratio_labels = []

for jj in range(1,6):
    for kk in range(1,6):
        if (jj/kk) not in ratios and (jj/kk) != 1:
            p_other = np.arange(((jj/kk)*p_koi3678)/1.05, ((jj/kk)*p_koi3678)*1.05, .1)
            ratios.append(jj/kk)
            ratio_labels.append(str(jj) + ':' +  str(kk))
            mmr_periods.append([p_other, MMR_super_period(p_koi3678, p_other, jj, kk)])
```

```python
plt.figure(figsize=[18,10])
for ii in range(0, len(mmr_periods)):
    p_other, super_period = mmr_periods[ii]
    ratio = ratios[ii]
    
    
    plt.plot(np.log10(p_other/p_koi3678), np.log10(super_period), color='k', lw=3)
    plt.axvline(np.log10(ratio), 0, 10, color='k', ls='dashed', lw=1)
    
plt.ylim(1.5,8)
plt.xlabel(r"(P$_\mathrm{KOI-3678c}$ : P$_\mathrm{KOI-3678b}$) in log$_{10}$ space", fontsize=36)
plt.ylabel(r"log$_{10}$(P$_\mathrm{TTV}$ [days])", fontsize=36)
plt.text(np.log10(1/5.5), 7.3, r'KOI-3678 expected TTV periods', fontsize=36, weight='bold')
plt.text(np.log10(1/5.5), 6.9, r'for a hidden planet that is < 5$\%$', fontsize=36)
plt.text(np.log10(1/5.5), 6.5, r'from orbital period resonance', fontsize=36)
plt.axhline(np.log10(415.18), 0, 800, color = 'k', lw=4, ls='--', label = r"Observed P$_\mathrm{TTV}$: 2.58 cycles = 415.08 days")

#plt.semilogy()
#plt.semilogx()


plt.xticks(ticks=np.log10(np.array(ratios)), labels=ratio_labels, fontsize = 18, rotation = 45)
plt.yticks(fontsize = 18)
plt.legend(loc=4, fontsize=27)
plt.tight_layout()
plt.savefig('koi3678_mmr_super_period.pdf')
```
