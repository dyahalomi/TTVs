```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider, Button

fig, ax = plt.subplots(3,1)
plt.subplots_adjust(left=0.2, bottom=0.1)

test_list = []
s0 = 0

t1 = np.arange(-1, 1, 0.1)
t2 = np.arange(-1, 1, 0.1)
t2 = np.arange(-1, 1, 0.1)

s1 = s0*np.ones(np.shape(t1))
s2 = s0*np.ones(np.shape(t2))
s3 = s0*np.ones(np.shape(t2))

l1, = ax[0].plot(s1, t1, lw=2, color='red')
l2, = ax[1].plot(s2, t2, lw=2, color='red')
l3, = ax[2].plot(s2, t2, lw=2, color='red')



ax[0].set_xlim(0,10)
ax[1].set_xlim(10,20)
ax[2].set_xlim(40,50)

axfreq = plt.axes([0.1, 0.25, 0.0225, 0.63], facecolor="lightblue")
sfreq = Slider(axfreq, 'time', 0, 50., valinit=s0, orientation="vertical")

def update(val):
    l1.set_xdata(val*np.ones(np.shape(t1)))
    l2.set_xdata(val*np.ones(np.shape(t2)))
    l3.set_xdata(val*np.ones(np.shape(t2)))

    fig.canvas.draw_idle()

sfreq.on_changed(update)

resetax = plt.axes([0.8, 0.025, 0.13, 0.04])
button = Button(resetax, 'save time', color=axcolor, hovercolor='0.975')

def save(event):
    test_list.append(sfreq.val)
    #print("Amplitude: ",  sfreq.val)

button.on_clicked(save)

plt.show()
```
